````python
## 회귀분석(Regression) 
## 11개의 독립변수 요소를 다 더해서 그 결과로 종속변수 qulity의 값이 결정된다
RFormula = 'quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates	+ alcohol + quality'
````

````python
## 회귀분석 결과
regrResult = ols(RFormula, data=dfTotalWine).fit()
````

````python
## 회귀분석 요약
regrResult.summary()
````
